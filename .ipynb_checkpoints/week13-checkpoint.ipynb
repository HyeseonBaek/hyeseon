```python
import csv
import matplotlib.pyplot as plt
f = open('seoul_.csv')
data = csv.reader(f)
next(data)
result1 = []
result2 = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '08':
            result1.append(float(row[-1]))
    if row[3] != '':
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '08':
            result2.append(float(row[3]))
plt.plot(result1,color='red')
plt.plot(result2,color='black')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul_.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```
