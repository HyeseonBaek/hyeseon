```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
max_temp = -999
max_date = ''
min_temp = 999
min_date = ''
sub_range = 0
sub_date = ''
for row in data :
    if row[-1] == '':
        row[-1] = -999
    row[-1] = float(row[-1])
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
    if row[3] == '':
        row[3] = 999
    row[3] = float(row[3])
    if min_temp > row[3]:
        min_date = row[0]
        min_temp = row[3]
    hot_temp = row[-1]
    cold_temp = row[3]
    sub_temp = hot_temp - cold_temp
    if sub_temp > sub_range:
        sub_range = sub_temp
        sub_date = row[0]    
f.close()
        
print('최저기온이 가장 낮은날')
print('날짜: '+min_date)
print('최고기온이 가장 높은날')
print('날짜: '+max_date)
print('일교차가 제일 큰 날')
print('날짜: '+sub_date)
```

```python
import csv
f = open('subway.csv')
data = csv.reader(f)
header = next(data)
max1_temp = -50000000
max1_name = ''
min1_temp = 50000000
min1_name = ''
for row in data:
    if row[4] == '':
        row[4] = -50000000
    row[4] = int(row[4])
    if max1_temp < row[4]:
        max1_temp = row[4]
        max1_name = row[3]
    if min1_temp > row[4]:
        min1_temp = row[4]
        min1_name = row[3]
f.close()

print('이용객이 가장 많은 역 순위')
print('1.',max1_name+' (',max1_temp,')')
print('이용객이 가장 적은 역 순위')
print('1.',min1_name+' (',min1_temp,')')
        
```
